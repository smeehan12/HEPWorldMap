# Get the Data
Here is where we use the INSPIRE-HEP API (https://github.com/inspirehep/rest-api-doc) to wget all of the data in a python digestible format

```python
import wget
import json
import os
import string# store all the json formatted files here
all_files=[]

alphabet_string = string.ascii_lowercase
alphabet_list = list(alphabet_string)
print(alphabet_list)

alphabet=alphabet_list
nget=250

for letter in alphabet:
    print("\nSearching: ",letter)
    
    # do first search to get Nhits
    url = 'https://inspirehep.net/api/institutions?sort=mostrecent&size='+str(nget)+'&page='+str(1)+'&q='+letter
    print("Getting: "+url)
    filename = wget.download(url)
    f = open(filename)
    data = json.load(f)
    f.close()
    os.system('rm institutions*')
    
    ntot = data['hits']['total']
    print("NTot: ",ntot)
    
    # get all of the data that contain that letter
    nquery = int(ntot/nget)+1
    print(nquery)
    
    for ipage in range(1,nquery+1):
        url = 'https://inspirehep.net/api/institutions?sort=mostrecent&size='+str(nget)+'&page='+str(ipage)+'&q='+letter
        print("Getting: "+url)
        filename = wget.download(url)
        f = open(filename)
        data = json.load(f)
        f.close()
        os.system('rm institutions*')
        hits = data['hits']['hits']
        print("NGot: ",len(hits))
        if len(hits)>0:
            all_files.append(data)
    
```

# Organize the data
Sift through the data to extract the number of publications and country code for the regional affiliation

Note here that I remove duplicate institutions, but I don't think this is entirely correct because they have different publication records

```python
dict_country_pubs={}
set_inst_id=set()

for data in all_files:
    hits = data['hits']['hits']
    print(len(hits))
    
    for inst in hits:
        # ID
        inst_id   = int(inst['id'])
        print(inst_id)
        
        # COUNTRY CODE
        if 'addresses' in inst['metadata']:
            if 'country_code' in inst['metadata']['addresses'][-1]:
                inst_code = str(inst['metadata']['addresses'][-1]['country_code']).lower()
            elif 'state' in inst['metadata']['addresses'][0]:
                country = str(inst['metadata']['addresses'][0]['state'])
                print(country)
                if country=="Benin":
                    inst_code = 'bj'
            else:
                city = str(inst['metadata']['addresses'][0]['cities'][0])
                if city=="Hanoi":
                    inst_code = 'vn'
        else:
            if 'UK' in inst['metadata']['ICN'][0]:
                inst_code = 'uk'
        
        # N PUBLICATIONS
        inst_pubs = inst['metadata']['number_of_papers']
        
        if inst_id in set_inst_id:
            print("WARNING: Duplicated Record ... skipping to add")
            continue
            
        print("noting that we are using this institute")
        set_inst_id.add(inst_id)
        
        print("adding to country spot")
        if inst_code not in dict_country_pubs:
            dict_country_pubs[inst_code] = inst_pubs
        else:
            dict_country_pubs[inst_code] += inst_pubs
```

# Plot the Data
Here we use the pygal library (https://www.pygal.org/) to make a pretty world map

```python
# import pygal library
import pygal
  
# create a world map
worldmap =  pygal.maps.world.World()
  
# set the title of the map
worldmap.title = 'Countries'
  
# adding the countries
worldmap.add('Number of Publications', dict_country_pubs)
  
# save into the file
worldmap.render_in_browser()
worldmap.render_to_file('worldmap.svg')
worldmap.render_to_png('worldmap.png')
  
print("Success")
```
